```python
# Номер 1
def factorize(n):
    L = []
    for i in range(2,round(n*0.5)+1):
        while n % i == 0 :
            L.append(i)
            n //= i
        if n ==1:
            break
    if n != 1:
        L.append(n)
    print(L)
n = int(input())
factorize(n)
```

```python
#Номер 2
import copy
def solve(n):
    if n == 0: return [[]]
    small_solve = solve(n-1)
    return [main_solution+[(n,i+1)]
        for i in range(8)
            for main_solution in small_solve
                if not attack(i+1, main_solution)]

def attack(column, queens): 
    left = right = column
    for r, c in reversed(queens):
        left, right = left-1, right+1
        if c in (left, column, right):
            return True
    return False

def queen():    
    L = []
    Lmatrix = []
    Answer = solve(8)
    for i in range(8):
        L.append('.')
    for i in range(8):
        Lmatrix.append(L[:])
    for row in Answer:
        Chess = copy.deepcopy(Lmatrix) # копирует без ссылок
        for Tuple in row:
            Chess[Tuple[0]-1][Tuple[1]-1] = '♕'
        for elem in Chess:
            print(*elem)
        print('\n')
queen()
```

```python
#Задача 3
def frequencies(filename):   
    file = open(filename,'r', encoding = 'UTF-8')
    s = file.read()
    s = s.lower()
    D = {}
    i = 0
    for symbol in s:
        if symbol.islower() and not ('a'<=symbol<='z'):
            i +=1
            D[symbol] = D.get(symbol,0) + 1 #обращаемся по индексу - символу. Get - взять значение из словаря по ключу. 0 - если ключ не найден, вернем число
    for key in D:
        D[key] = (D[key]/i)*100
        print('%s: %.2f%%' % (key, D[key]))
    file.close()
    return D
frequencies('tolstoi.txt')
```

```python
#Задача 4
def decipher(filename, enc):
    Dnew = {}
    EncDict = frequencies(enc)
    FileDict = frequencies(filename)
    
    EncList = [ [k,v] for k, v in EncDict.items() ] #для каждой пары делаем список из двух эл-ов
    FileList = [ [k,v] for k, v in FileDict.items() ]
    EncList.sort(key=lambda x: x[1], reverse = True)
    FileList.sort(key=lambda x: x[1], reverse = True)
    
    for i in range(len(EncList)):
        Dnew[EncList[i][0]] = FileList[i][0]
    
    encfile = open(enc,'r', encoding = 'UTF-8')
    s = encfile.read()
    encfile.close()
    s = s.lower()
    for elem in s:
        if elem.islower() and not ('a'<=elem<='z'):
            print(Dnew[elem], end = '')
        else:
            print(elem, end = '')
decipher('tolstoi.txt','tolstoi.enc')
```

```python
import re
import urllib.request
from bs4 import BeautifulSoup
url = 'http://sophist.hse.ru/exes/tables/CPI_Y_CHI.htm'
html = urllib.request.urlopen(url).read()
soup = BeautifulSoup(html, 'html.parser')
cpi = soup.find_all('td', align = 'right')
i = 0
date = []
percent = []
for elem in cpi:
    #print(str(elem).find('>'))
    s = str(elem)
    s = s.replace(',','.')
    num_index_beg = ((s.find('>', 1 + s.find('>')))) + 1
    num_index_end = s.find('<', num_index_beg)
    num = float(s[num_index_beg:num_index_end])
    
    if (i % 2 == 0):
        date.append(int(num))
    else:
        percent.append(num - 100)
    i += 1


index = date.index(2000)
x = Data(percent[index:])
print(x)
```
